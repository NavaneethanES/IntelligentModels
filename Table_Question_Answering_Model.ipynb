```python
from transformers import pipeline
import pandas as pd
```

```python
tqa=pipeline(task="table-question-answering",model="google/tapas-base-finetuned-wtq")
```

```python
table=pd.read_csv("/content/netflix_titles.csv",encoding='latin-1')
```

```python
table=table.astype(str)
```

```python
table=table
```

```python
table.drop(table.columns[11:],axis=1,inplace=True)
```

```python
table.drop(table.index[10:],inplace=True)
```

```python
table
```

```python
query= "country for kota factory ?"
```

```python
print(tqa(table=table,query=query)["answer"])
```
